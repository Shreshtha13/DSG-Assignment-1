```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import OLSInfluence

%matplotlib inline
plt.style.use('ggplot')
```

```python
# creating vector x with values from N(0,1)
np.random.seed(0)
x=np.random.randn(100)
eps=0.25*np.random.randn(100)
y=-1+0.5*x+eps
print(len(y))
print('beta0 and beta1 are -1 and 0.5 respectively')
```

```python
# scatterplot
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(x,y,facecolors='none',edgecolors='b',label='data')
# Least Squares model
X=sm.add_constant(x)
model=sm.OLS(y,X)
est=model.fit()
# Least Squares Line
ax.plot(x,est.fittedvalues,color='red',label='Best Fit')
ax.plot(x,-1+0.5*x,color='k',label='Population Regression Line')
ax.legend(loc='best')

```

```python
x_sq=x**2
X=np.concatenate((X,x_sq.reshape(len(x),1)),axis=1)
model_poly=sm.OLS(y,X)
est_poly=model_poly.fit()
print(est_poly.summary())
```

```python
# decreasing noise in the model
np.random.seed(0)
x_lessN=np.random.randn(100)
eps_lessN=0.05*np.random.randn(100)
y_lessN=-1+0.5*x_lessN + eps_lessN
X_less=sm.add_constant(x_lessN)
# print(X_less).
model_less=sm.OLS(y_lessN,X_less)
est_less=model_less.fit()
fig,axl=plt.subplots(figsize=(10,10))
axl.scatter(x_lessN,y_lessN,label='data')
axl.plot(x_lessN,est_less.fittedvalues,color='k',label='ols')
axl.legend()
print(est_less.summary())
# beta0 - [-1.007 to -0.986]
# beta1 - [0.496 to 0.516]
```

```python
# increasing noise in the model
np.random.seed(0)
x_moreN=np.random.randn(100)
eps_moreN=0.5*np.random.randn(100)
y_moreN=-1+0.5*x_moreN + eps_moreN
X_more=sm.add_constant(x_moreN)
# print(X_less).
model_more=sm.OLS(y_moreN,X_more)
est_more=model_more.fit()
fig,axl=plt.subplots(figsize=(10,10))
axl.scatter(x_moreN,y_moreN,label='data')
axl.plot(x_moreN,est_more.fittedvalues,color='k',label='ols')
axl.legend()
print(est_more.summary())
# beta0 - [-1.066 to -0.859]
# beta1 - [0.455 to 0.660]
```
